# Fundamentos de Procesamineto Digital de Señales

- Las señales son las manifestaciones físicas de procesos naturales o artificiales de muy diferentes natualezas
- Caracteristicas comunes
  - Son funciones de una o mas variables independientes
  - Contiene informacion acerca del comportamiento o la naturaleza del fenomeno físico en cuestion

```python
# Importamos la biblioteca para dibujar
import matplotlib.pyplot as plt
# Importamos la biblioteca para análisis numerico
import numpy as np
```

## Señales continuas
Como vemos las señales continuas en el tiempo o analigicas contiene toda la informacion son señales que nunca se muestrean , es la señal tal y como viene de la naturaleza.

```python
# Dibujaremos la señal analogica 
t = np.arange(0,np.pi*3,.1)
y = np.sin(t)
fig, ax = plt.subplots()
ax.plot(t, y)

ax.set(xlabel='tiempo (s)', ylabel='Voltaje (mV)',
       title='Señal en el tiempo continuo x(t)')
ax.grid()
plt.show()
```

## Señales discretas 
Son señales que se muestrean de la señal en el tiempo continuo , se van tomando muestras con una frecuencia determinada , osea nunca se tiene todos los valores solo una parte de ellos.

```python
# Dibujaremos la señal analogica 
x = np.arange(0,np.pi*3,.5)
y = np.sin(x)
n = np.arange(len(x))
fig, ax = plt.subplots()
ax.plot(n, y,'-o')

ax.set(xlabel='Muestras n', ylabel='Voltaje (mV)',
       title='Señal digital')
ax.grid()
plt.show()
```

Uno de los problemas mas comunes es que aun teniendo todos la datos de fenomeno, en los formatos analogicos exite ruido lo ilustraremos en la parte de abajo , a pesar de que tiene mucha informacion mucha de esa informacion es ruido. Esa es la razon por la cual un VHS es de menor calidad que un BlueRay , o la radio analogica tiene menor calidad que la radio digital o formatos como los de Spotify.

```python
# Dibujaremos la señal analogica 
t = np.arange(0,np.pi*3,.1)
ruido = []
for _ in range(len(t)):
  ruido.append(np.random.random_sample())
y = np.sin(t) + ruido
y_sin_r = np.sin(t)
##### Primera grafica
fig, axs = plt.subplots(2,1)
axs[0].plot(t, y)
axs[0].set(xlabel='tiempo (s)', ylabel='Voltaje (mV)',
       title='Señal en el tiempo continuo x(t) con ruido')
axs[0].grid()
##### Segunda grafica
axs[1].plot(t, y_sin_r)
axs[1].set(xlabel='tiempo (s)', ylabel='Voltaje (mV)',
       title='Señal en el tiempo continuo x(t) con ruido')
axs[1].grid()
plt.show()
```

En el munodo digital hay un efecto parecido al fuido y se llama Alliasing
Es cuando el sistema de muestreo no es lo sificientemente poderoso para poder calpurar toda la infomacion necesaria. Y pareciece que no se mueven la cosas.
Esta relacionado con la frecuencia con la que los humanos y los sistemas capturas las cosas y producen un efecto como el siguiente.


```python

%matplotlib inline
# Dibujaremos la señal analogica 
x = np.arange(0,np.pi*12,.5)
y = np.sin(x)

####
x_ = np.arange(0,np.pi*12,.5)
y_ = np.sin(x_*.4)
n = np.arange(len(x))
fig, ax = plt.subplots()
ax.stem(n, y,':','-o',use_line_collection=True)

ax.set(xlabel='Muestras n', ylabel='Voltaje (mV)',
       title='Señal digital')
ax.stem(n, y_,':','r',linefmt='grey',use_line_collection=True)
ax.grid()
plt.show()
```

## Señales elementales

```python
# Dibujaremos la señal analogica 
x = np.arange(0,np.pi*3,.45)
y = np.sin(x)
n = np.arange(len(x))
fig, ax = plt.subplots()
ax.stem(n, y,':',use_line_collection=True,label = r'$X(n)$')
ax.plot(n,y,'r',label = r'$X_a(t)$')
ax.set(xlabel='Muestras n', ylabel='Voltaje (mV)',
       title='Señal digital')

for i, txt in enumerate(y):
    plt.annotate(str(round(txt,2)), (n[i],y[i]))

plt.legend(loc="lower right")
ax.grid()
plt.show()
```

La señal $X(n)$ esta definida para los intervalos  $ -\infty	< n < \infty $

Si $X(n)$ se ha obtenido muestreando una señal analógica $X_a(t)$

Entonces 

$X(n) = X_a(nT)$

Se ha visto que $x(n)$ es una función definida para n entero.
La figura presenta un ejemplo de representación gráfica de una señal de este tipo
Como consecuencia
$x(n) =\{..., 0 ,0, 1, 3, 2 ,1,0    \}$
La mayoria de la veces vamos a encontrar la definicion de la funcion de 
la siguiente forma
\begin{equation}
 f(n) = \left \{ \begin{matrix} 1 & \mbox{para }n\mbox = 1
\\ 5-n & \mbox{para }\mbox2<n\mbox{ es }\end{matrix}\right. 
\end{equation}


```python
# Dibujaremos la señal analogica 
x = np.arange(0,7,1)
y = [0,0,1,3,2,1,0]
n = np.arange(len(x))
fig, ax = plt.subplots()
ax.stem(n, y,':',use_line_collection=True,label = r'$X(n)$')
plt.axis([-1, 8, 0, 5])
for i, txt in enumerate(y):
    plt.annotate(str(round(txt,2)), (n[i],y[i]+.2))

plt.legend(loc="upper right")
ax.grid()
plt.show()
```

## Graficacion de varias funciones
La rampa unitaria
$ u_r(n) = \sum_{i=\infty}^{3} u(i-1)$ 


lo que resulta en la figura
\begin{equation}
 u_r(n) = \left \{ \begin{matrix} 0 & \mbox{ para } n\mbox < 0
\\n & \mbox   {para } \mbox n>0\mbox{  }\end{matrix}\right. 
\end{equation}

```python
# Dibujaremos una selal discreta rampa
n = np.arange(0,10)
y_r_n =  [i if i>0 else 0 for i in n] 
fig , ax = plt.subplots()
ax.stem(n,y_r_n,':',use_line_collection=True,label=r'$u_(n)$')
plt.axis([-1, 11, 0, 11])
for i, txt in enumerate(y_r_n):
    plt.annotate(str(round(txt,2)), (n[i],y_r_n[i]+.2))

plt.legend(loc="upper right")
ax.grid()
plt.show()
```

Una señal $X(n)$ se puede desplazar en el tiempo reemplazando la variable independiente $n$ por $n-k$ 

```python
def rampa(k):
  n = np.arange(0,10) + k
  print(n)
  y_r_n =  [(i) if i>0 else 0 for i in n] 
  fig , ax = plt.subplots()
  ax.stem(n,y_r_n,':',use_line_collection=True,label=r'$u_(n)$')
  plt.axis([min(n)-1, 11, 0, 11])
  for i, txt in enumerate(y_r_n):
      plt.annotate(str(round(txt,2)), (n[i],y_r_n[i]+.2))

  plt.legend(loc="upper right")
  ax.grid()
  plt.show()
rampa(-3)
```
